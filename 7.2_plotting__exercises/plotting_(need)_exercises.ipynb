## 1 Fill in the Blanks

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err, # Setting value of error bars
             color="maroon", linestyle="dashed", # Setting style of the line plot
             marker="o", markerfacecolor="yellow", # Setting style of individual points
             ecolor="black",capsize=5, # Setting style of error bars
             label="Y values with errors") # Display text for the legend
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc="lower right") # Location of the legend
plt.show()
```

## 2 A simple mathematical plot

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 50, 300) # More points for a smoother curve
exp_x = np.exp(-x/10)
sin_x = np.sin(x)
f_x = exp_x * sin_x 
plt.grid(alpha=.25) # Plotting the background grid

plt.plot(x, f_x)
plt.title(r"$y = e^{-x/10}sin(x) $", fontsize="20") 
plt.xlabel("x", color="darkorange", fontsize="15")
plt.ylabel("y", color="darkorange", fontsize="15", rotation="horizontal") # y label needs to be rotated
plt.ylim(-1,1) # Changing the values on the y-axis
```
